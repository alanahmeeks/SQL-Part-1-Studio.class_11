Question 5: How many books where released in the first decade of 2000?

```sql
-- return number of books published in the first decade of 2000
select count(*) as 'books_from_2000s'
from books 
where original_publication_year between 2000 and 2009
```

Question 6: How many book titles contain the word Happy?

```sql
-- return number of books that contain the word Happy
SELECT count(*) AS 'Happy Books'
FROM books 
WHERE title LIKE '%happy%'
```

## Part 2. Write your own question and write a query to answer it.

What is the total average rating for each Author from all books

```sql
-- Add your querry here
SELECT AVG(average_rating) AS 'total_avg_rating', authors
FROM books 
GROUP BY authors
ORDER BY total_avg_rating DESC
```

add to show total number of books for each author

```sql
-- Add your querry here
SELECT AVG(average_rating) AS 'total_avg_rating', authors, COUNT(title) AS 'number_of_books'
FROM books 
GROUP BY authors
ORDER BY total_avg_rating DESC
```
